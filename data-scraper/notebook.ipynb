```python

# Python program to demonstrate
# Conversion of JSON data to
# dictionary
 
 
# importing the module
import json
import re
import pandas as pd
import numpy as np
from IPython.display import display
from utils import *

def json_to_df(file_name='json_test.txt',cutoff = 5):
# Opening JSON file
    with open(file_name) as f:
        data = f.read().replace("\\","/").replace('""','","')
    with open(file_name+"converted","w") as d:
        d.write(data)
    print("Data type before reconstruction : ", type(data))
    # reconstructing the data as a dictionary
    js = json.loads(data,strict=False)
    df = pd.read_json(file_name+"converted")

    print("Data type after reconstruction : ", type(js))
    display(df.head(5))
    comp_df= pd.DataFrame(df.drop(['project_link','project_title','project_description','tools'], axis = 1))
    print(len(df))
    display(comp_df.head(2))
    list_of_zeros = [0]*len(df)
    #print(list_of_zeros)
    #display(list(df["project_id"]))
    components = read_columns ("components.txt")
    df_header=["project_id"]+list(components)
    df2 = pd.DataFrame(0, index=np.arange(len(df)), columns=df_header)
    df2['project_id'] = list(df["project_id"])
    display(df2.head(2))
    df2=df2.copy()
    df2.info()
    for i in range (0,len(df)):
        for j in range(0,len(df.iloc[i]["components"])):
            comp_index=df.iloc[i]["components"][j]
            df2.loc[i,comp_index]=1
    display(df2.head(2))
    for i in components:
        sum = df2[i].sum()
        if (sum<cutoff):
            #print('sum : ',sum)
            df2=df2.drop([i], axis = 1)
    df2.info()
    display(df2.head(10))
    return df2
    
        



    




df2 = json_to_df('all_project_detail_final_test.txt',cutoff=2)
```

```python
# similarity
from matplotlib.pyplot import axis
from scipy.spatial.distance import cosine
import math
import random

def get_similar_projects (project, df2, cutoff = 0.1):
    df3=df2
    scores = []
    for row in range(0, len(df3)):
        sim_score= 1 - cosine(project,df3.loc[[row]])
        scores.append(sim_score)
    df3['scores']= scores
    for row in range(0, len(df3)):
        if (df3.loc[row,'scores'] <= cutoff or df3.loc[row,'scores'] == 0.0 or math.isnan(df3.loc[row,'scores'])):
            df3.drop(row, axis = 0 , inplace = True)
    df3.sort_values('scores', ascending=False, inplace = True)
    for i in list(df3.columns.values):
        if(df3[i].sum()==0):
            df3.drop(i, axis = 1, inplace = True)
    
    return df3


def generate_input(df):
    input_list= []
    for i in range (0,len(df.columns)):
        input_list.append(random.randint(0,1))
    return input_list

similar_list = get_similar_projects (generate_input(df2.drop(['project_id'],axis=1)), df2.drop(['project_id'],axis=1), cutoff=0.25)
display(similar_list)
#print(df2.loc[140])
```
