```python
"""
transform a dataframe into a suprire formatted dataframe
"""

def df_to_surprise_df(input_file='project_df_cut_5_5_test.csv', output_file='out.csv'):

    comp_df = pd.read_csv(input_file, index_col=0)
    df_header=['userID','itemID','rating']
    df2 = pd.DataFrame(columns=df_header)
    components = comp_df.columns
    projects = comp_df.index
    list_of_zeros = [0]*(len(components)*len(projects))
    index = pd.MultiIndex.from_product([projects, components], names = ["userID", "itemID"])
    df3= pd.DataFrame(index = index)
    df3['rating']=list_of_zeros
    for i in comp_df.index:
        for j in comp_df.columns:
            if comp_df.loc[i,j]==1:
                df3.loc[i,j]=1 
    display(df3)
    df3.to_csv(output_file)
    return df3
```
